### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
#There is the  explicit and the implicit way
#The explicit way Explicitly create Figures and Axes, and call methods on them and it is called(the object oriented style). 
#The implicit way Rely on pyplot to implicitly create and manage the Figures and Axes, and use pyplot functions for plotting,
#and it is called(the pyplot style).
#
# Even though the Matplotlib's documentation and examples use both the object oriented and the pyplot styles.
# In general, they recommend using the object oriented style, particularly for complicated plots, and functions and scripts
# that are intended to be reused as part of a larger project. 



```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
#The Figure keeps track of all the child Axes, a group of Artists (titles, figure legends, colorbars, etc.),
# and even nested subfigures.we can create figures through different functions.

#An Axes is an Artist attached to a Figure that contains a region for plotting data.
#The Axes methods are the primary interface for configuring most parts of your plot. 

#Axis are the objects that set the scale and limits and generate ticks (the marks on the Axis),
# and ticklabels (strings labeling the ticks).

#Basically, everything visible on the Figure is an Artist (even Figure, Axes, and Axis objects).
# Most Artists are tied to an Axes ,such an Artist cannot be shared by multiple Axes, or moved from one to another.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
def f(x):
   return x**2
x = np.linspace(-4,4,50)

#the object oriented style
fig, ax = plt.subplots()
ax.plot(x,f(x))
ax.set_title(r'My Function $y = x^2$')
ax.set_xlabel(r'X axis')
ax.set_ylabel(r'Y axis')

#the pyplot style
plt.figure()
plt.title("My Function $y = x^2$")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x,f(x))
plt.show()
```
